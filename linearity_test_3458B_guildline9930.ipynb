```python
import numpy as np
import matplotlib.pyplot as plt

# Array from .csv

arra=np.genfromtxt('csv/linearity_test_3458B_guildline9930_2025-09-22_11-56-09.csv',delimiter=',',skip_header=1)

# Columns to lists
G9930 = [col[0] for col in arra]
K3458B = [col[1] for col in arra]
PPMdeviationa = [(K3458B[x]-G9930[x])/(10/1000000) for x in range(len(K3458B))]

figa, axa = plt.subplots()
#axa.plot(F5700A, PPMdeviationa, label='absolute')
axa.set_xlabel('Unmodified Guildline 9930 Vset')
axa.set_ylabel('PPM of 10V')
axa.set_title("Best-straight-line-INL of 3458A 1VDC vs. Guildline 9930")
figa.set_size_inches(16, 9)
figa.patch.set_facecolor('white')

# Fit a polynomial function of degree 1
termsa = np.polyfit(G9930, PPMdeviationa, 1)

# Create a function with the terms we just generated
polyfunctiona = np.poly1d(termsa)

bestfita = [PPMdeviationa[x]-polyfunctiona(G9930[x]) for x in range(len(K3458B))]

# Plot a little test
axa.plot(G9930, bestfita, label='Keysight 3458B Run 1')








arra=np.genfromtxt('csv/linearity_test_3458B_guildline9930_2025-09-22_14-12-02.csv',delimiter=',',skip_header=1)
G9930 = [col[0] for col in arra]
K3458B = [col[1] for col in arra]
PPMdeviationa = [(K3458B[x]-G9930[x])/(10/1000000) for x in range(len(K3458B))]
termsa = np.polyfit(G9930, PPMdeviationa, 1)
polyfunctiona = np.poly1d(termsa)
bestfita = [PPMdeviationa[x]-polyfunctiona(G9930[x]) for x in range(len(K3458B))]
axa.plot(G9930, bestfita, label='Keysight 3458B Run 4')

arra=np.genfromtxt('csv/linearity_test_3458B_guildline9930_2025-09-22_14-22-07.csv',delimiter=',',skip_header=1)
G9930 = [col[0] for col in arra]
K3458B = [col[1] for col in arra]
PPMdeviationa = [(K3458B[x]-G9930[x])/(10/1000000) for x in range(len(K3458B))]
termsa = np.polyfit(G9930, PPMdeviationa, 1)
polyfunctiona = np.poly1d(termsa)
bestfita = [PPMdeviationa[x]-polyfunctiona(G9930[x]) for x in range(len(K3458B))]
axa.plot(G9930, bestfita, label='Keysight 3458B Run 5')





arra=np.genfromtxt('csv/linearity_test_3458A_guildline9930_2025-09-22_15-04-42.csv',delimiter=',',skip_header=1)
G9930 = [col[0] for col in arra]
K3458B = [col[1] for col in arra]
PPMdeviationa = [(K3458B[x]-G9930[x])/(10/1000000) for x in range(len(K3458B))]
termsa = np.polyfit(G9930, PPMdeviationa, 1)
polyfunctiona = np.poly1d(termsa)
bestfita = [PPMdeviationa[x]-polyfunctiona(G9930[x]) for x in range(len(K3458B))]
axa.plot(G9930, bestfita, label=' HP 3458A Run 2')

arra=np.genfromtxt('csv/linearity_test_3458A_guildline9930_2025-09-22_15-17-01.csv',delimiter=',',skip_header=1)
G9930 = [col[0] for col in arra]
K3458B = [col[1] for col in arra]
PPMdeviationa = [(K3458B[x]-G9930[x])/(10/1000000) for x in range(len(K3458B))]
termsa = np.polyfit(G9930, PPMdeviationa, 1)
polyfunctiona = np.poly1d(termsa)
bestfita = [PPMdeviationa[x]-polyfunctiona(G9930[x]) for x in range(len(K3458B))]
axa.plot(G9930, bestfita, label=' HP 3458A Run 3')


axa.legend()
```
